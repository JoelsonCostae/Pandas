#Criar dataframe - df

* Leitura do arquivo de dados:

* Se não escolhermos uma coluna, irá criar o rotulo automaticamente, da posição 0 até a ultima linha-1, porém, podemos escolher uma coluna como rotulo, apenas para LOCALIZAÇÃO[linha,coluna]

* Indice X Rotulo : o rotulo pode ser modificado e mostra qual a localização de determinada coisa na tabela [linha,tabela], não tem como localizar pelo -1 para percorrer de tras pra frente como o indice, e o indice não pode ser modificado

* Para mostrar o dataframe basta chamar o nome dele  na ultima linha do codigo (sem print)

AAAA - MM - DD
