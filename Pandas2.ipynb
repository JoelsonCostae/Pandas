#**Validação de dados**

1. Passar o Schema para ele


> Podemos escolher o nome que quiser, mas por convenção, chamamos de Schema

* Validar o tipo de dados de cada coluna

schema = pa.DataFrameSchema(
  columns = {
  "nomecoluna" : pa.Column(**pa.tipo de dados**),
  ...
})

* Validar a quantidade de caracteres

"nomecoluna" : pa.Column(pa.tipo,**pa.Check.str_lenght(min,max)**)

* Aceitar valores nulos:

**nullable = True**

* Verificar o formato das horas

**Horas:**

Possibilidade 1 : Antes do | [0-1]?[0-9]
- O primeiro digito aceita de 0 a 1 [0-1]
- O segundo digito aceita de 0 a 9 [0-9]

** O ? é como uma pergunta : O primeiro digito é 0 ou 1? Se sim, o segunto tem que estar entre 0 e 9. SENÃO (|), o primeiro digito será 2 e o segundo estrá entre 0 e 3**

Possibilidade 2 : Depois do | [2][0-3]
- O primeiro digito aceita apenas o 2 [2]
- O segundo digito aceita apenas de 0 a 3 [0-3]

**Minutos e Segundos:**

- Primeiro digito entre 0 e 5
- Segundo digito entre 0 e 9

- Separados por dois pontos ( : )

`pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9]):([0-5][0-9])$')`

* Inicia com r'^
* Finaliza com $'

```python
schema = pa.DataFrameSchema(
    columns = {
        #Verifica se a coluna é do tipo passado
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2" : pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        #Valida a quantidade de caracteres 
        "ocorrencia_uf" : pa.Column(pa.String,pa.Check.str_length(2,2),nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String,pa.Check.str_length(4,4),nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9]):([0-5][0-9])$'),nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    }

)
```

**Fazer a validação**
* Se aparecer o Dataframe é pq ta validado

```python
schema.validate(df)
```
